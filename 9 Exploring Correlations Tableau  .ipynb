### 2b Correlation Linear Regression (Low Household Income participants cleaned)

```python
# Create a scatterplot using matplotlib - how the chosen variables plot against each other.

Data_rlhi.plot(x = 'Equipment', y='Social Life Frequency', style='o') 
plt.title('Equipment x Social Life')  
plt.xlabel('Equipment')  
plt.ylabel('Social Life Frequency')  
plt.show()
```

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = Data_rlhi['Equipment'].values.reshape(-1,1)
y = Data_rlhi['Social Life Frequency'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Create a regression object.

regression = LinearRegression()
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
# Predict the values of y using X.

y_predicted = regression.predict(X_test)
```

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Equipment x Social Life')
plot_test.xlabel('Equipment')
plot_test.ylabel('Social Life Frequency')
plot_test.show()
```
